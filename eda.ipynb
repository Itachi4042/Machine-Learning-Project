```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 

```

```python
df=pd.read_csv("C:\\Users\\SHIVAM AGARWAL\\OneDrive\\Downloads\\upi_transactions_2024.csv")
df.head()
```

```python
df.info()
```

```python
cat_cols=['transaction type', 'merchant_category', 'transaction_status', 'sender_state', 'receiver_bank', 'sender_bank', 'device_type','network_type', 'day_of_week']
for i in cat_cols:
    print(df.value_counts(i),"\n")
```

```python
def plot_bar(col):
    plt.figure()
    value_counts = df[col].value_counts().head(10)  # Top 10 only to avoid clutter
    sns.barplot(x=value_counts.index, y=value_counts.values, palette='pastel')
    plt.title(f'{col} Distribution')
    plt.xlabel(col.replace('_', ' ').title())
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Plot all the categorical columns
for col in cat_cols:
    plot_bar(col)
```

```python
import numpy as np

# Create a new log-transformed column
df['log_amount'] = np.log1p(df['amount (INR)'])

# Plot log-transformed distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['log_amount'], kde=True, bins=50, color='mediumseagreen')
plt.title('Log-Transformed Distribution of Transaction Amount')
plt.xlabel('log(1 + Amount)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['hour_of_day'], bins=24, color='orange', kde=False)
plt.title('Transaction Count by Hour of Day')
plt.xlabel('Hour (0 to 23)')
plt.ylabel('Number of Transactions')
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x='fraud_flag', palette='Set1')
plt.title('Fraud vs Non-Fraud Transactions')
plt.xlabel('Fraud Flag (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Transaction Count')

# Add count labels on top of bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:,}',             
                (p.get_x() + p.get_width() / 2, height), 
                ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='sender_age_group', y='amount (INR)', palette='viridis', ci=None)
plt.title('Average Transaction Amount by Sender Age Group')
plt.xlabel('Sender Age Group')
plt.ylabel('Average Amount (INR)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='device_type', palette='Set2')
plt.title('Transaction Count by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Transaction Count')

# Add labels
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()/2, p.get_height()), 
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
success_rate = df.groupby('network_type')['transaction_status'].value_counts(normalize=True).unstack()['SUCCESS'] * 100

plt.figure(figsize=(7, 4))
sns.barplot(x=success_rate.index, y=success_rate.values, palette='coolwarm')
plt.title('Success Rate per Network Type')
plt.xlabel('Network Type')
plt.ylabel('Success Rate (%)')

# Add percentage labels
for i, v in enumerate(success_rate.values):
    plt.text(i, v + 0.5, f'{v:.1f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
fraud_by_bank = df[df['fraud_flag'] == 1]['sender_bank'].value_counts().nlargest(10)

plt.figure(figsize=(8, 5))
sns.barplot(x=fraud_by_bank.values, y=fraud_by_bank.index, palette='Reds_r')
plt.title('Top 10 Banks with Most Fraudulent Transactions')
plt.xlabel('Fraud Count')
plt.ylabel('Sender Bank Name')
plt.tight_layout()
plt.show()
```

```python
# Select numerical columns
numerical_cols = ['amount (INR)', 'hour_of_day', 'fraud_flag']
corr_matrix = df[numerical_cols].corr()
corr_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Convert to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract components for time-based analysis
df['hour_of_day'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek  # Monday = 0, Sunday = 6
df['date'] = df['timestamp'].dt.date
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='hour_of_day', palette='viridis')
plt.title('Transactions by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Transaction Count')
plt.tight_layout()
plt.show()
```

```python
# Add a weekend flag (0 = Weekday, 1 = Weekend)
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

plt.figure(figsize=(5, 4))
sns.countplot(data=df, x='is_weekend', palette='coolwarm')
plt.xticks([0, 1], ['Weekday', 'Weekend'])
plt.title('Transactions: Weekday vs Weekend')
plt.xlabel('')
plt.ylabel('Transaction Count')
plt.tight_layout()
plt.show()
```

```python
# Sender bank
top_banks = df['sender_bank'].value_counts().head(10).index
plt.figure(figsize=(10, 4))
sns.countplot(data=df[df['sender_bank'].isin(top_banks)], x='sender_bank', hue='fraud_flag', palette='cool')
plt.title('Top Sender Banks by Fraud Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
weekend_fraud = df.groupby('is_weekend')['fraud_flag'].sum()

plt.bar(['Weekday', 'Weekend'], weekend_fraud, color=['skyblue', 'orange'])
plt.title('Fraud Count on Weekdays vs Weekends')
plt.ylabel('Fraud Transactions')
plt.tight_layout()
plt.show()
```
